- Correlazione produzione/consumi
    - [https://www.nature.com/articles/s41598-020-68231-8.pdf](https://www.nature.com/articles/s41598-020-68231-8.pdf)

```python
GHG_by_country_and_culture = pd.read_excel("/content/drive/MyDrive/FOOTPLUS/GHG_by_country_and_culture.xlsx")
GHG_by_country_and_culture.head()
```

```python
production_by_country = production_by_country_and_culture[['Country', 'Total']]
production_by_country = production_by_country.rename(columns={'Total': "Production"})
GHG_by_country = GHG_by_country_and_culture[['Country', 'Total']]
GHG_by_country = GHG_by_country.rename(columns={'Total': "GHG"})
correlation = pd.merge(production_by_country, GHG_by_country, on='Country')
correlation.head()
```

```python
fig = px.scatter(correlation,
    x='Production', 
    y='GHG',
    hover_data=["Country"]
)
fig.update_yaxes(title_text="Greenhouse emissions")
fig.update_xaxes(title_text="Aquaculture production")

fig.show()
```

- Pesci pi eco friendly

```python
footprint_by_country_and_culture = GHG_by_country_and_culture.drop(['Total'], axis=1)
columns_to_divide = ['Bivalves ', 'Catfish', 'Cyprinids ', 'Freshwater fish,general', 'Indian major carps', 'Marine fish,general', 'Salmonids', 'Shrimps and prawns', 'Tilapias ']
footprint_by_country_and_culture[columns_to_divide] = footprint_by_country_and_culture[columns_to_divide]/production_by_country_and_culture[columns_to_divide]
footprint_by_country_and_culture = footprint_by_country_and_culture.replace([np.inf, -np.inf], np.nan)
footprint_by_country_and_culture = footprint_by_country_and_culture.fillna(0)
footprint_by_country_and_culture = footprint_by_country_and_culture.drop([10])
footprint_by_country_and_culture.head()
```

```python
fish_footprint = {}
for row in footprint_by_country_and_culture.iterrows():
    for fish in columns_to_divide:
        if row[1][fish] != 0.0:
            fish_footprint[row[1]['Country'] + ' '  + fish] = row[1][fish]
fish_footprint = pd.DataFrame.from_dict(fish_footprint, orient='index', columns=['footprint'])
fish_footprint.head()
```

```python
fig = px.bar(
    x=fish_footprint.index,
    y=fish_footprint['footprint'],
    title='Aquaculture fish footprint'
).update_xaxes(categoryorder="total ascending")

fig.show()
```

    - [https://link.springer.com/content/pdf/10.1007/s13280-018-1060-9.pdf](https://link.springer.com/content/pdf/10.1007/s13280-018-1060-9.pdf)

```python
aquaculture_meal = 18 # milion tons
GHG_fish = 5.4 # kgCO2/kg

total_GHG_meal = aquaculture_meal * GHG_fish
total_GHG_meal
```

```python
GHG_by_specie.loc[GHG_by_specie['Specie']=='Aquaculture', ['GHG']] += total_GHG_meal
```

```python
fig = px.bar(
    x=GHG_by_specie['Specie'],
    y=GHG_by_specie['GHG'],
    title='Milion tonnes of emitted GHG'
).update_xaxes(categoryorder="total ascending")

fig.show()
```
