```python
import pandas as pd
```

```python
df_CO2_fishing = pd.read_csv('../Preprocessed_data/Flourish_data/total_CO2_emissions_plot.csv')
df_CO2_fishing['total_CO2']=df_CO2_fishing['total_CO2'].astype(int)
df_CO2_fishing = df_CO2_fishing.rename(columns={'year': 'Year'})
df_CO2_fishing
```

```python
df_CO2_meat =  pd.read_csv('../Raw_Data/global_CO2_emission_agriculture.csv')
df_CO2_meat = df_CO2_meat.drop(columns=['Domain Code','Domain', 'Area Code', 'Area', 'Element Code', 'Element', 
                                        'Item Code', 'Item', 'Year Code', 'Unit', 'Flag', 'Flag Description', 'Note'])
df_CO2_meat['Value'] = df_CO2_meat['Value'].astype('uint64')
df_CO2_meat
```

```python
df_CO2_meat['Value']=df_CO2_meat['Value'] * 1000 #converting gigagrams to tons, to have the same unit measure between fisheries and meat
df_CO2_meat
```

```python
df_final =  pd.merge(df_CO2_meat, df_CO2_fishing, on="Year")
df_final
```

```python
df_final.to_csv('../Preprocessed_data/Flourish_data/fish_meat_CO2_comparison.csv')
```
